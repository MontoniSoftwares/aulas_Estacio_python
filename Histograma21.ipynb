```python
import numpy as np
import matplotlib.pyplot as plt

# Gera os 1000 pontos seguindo a distribuição normal
media = 20
desvio_padrao = 2
pontos = np.random.normal(media, desvio_padrao, 1000)

# Cria o histograma
plt.hist(pontos, bins=50)

# Define o título do gráfico
plt.title('Histograma de pontos seguindo distribuição normal')

# Define o rótulo do eixo x
plt.xlabel('Valores')

# Define o rótulo do eixo y
plt.ylabel('Frequência')

# Exibe o gráfico
plt.show()
```
