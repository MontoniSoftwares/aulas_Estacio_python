```python
import matplotlib.pyplot as plt

# Dados a serem plotados
x = ['A', 'B', 'C', 'D']
y = [10, 20, 30, 40]

# Cria o gráfico de barras
plt.bar(x, y)

# Define o título do gráfico
plt.title('Exemplo de gráfico de barras')

# Define o rótulo do eixo x
plt.xlabel('Eixo x')

# Define o rótulo do eixo y
plt.ylabel('Eixo y')

# Exibe o gráfico
plt.show()
```
